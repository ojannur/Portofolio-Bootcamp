2. Berasal dari agensi travel mana saja nasabah yang klaim asuransi?

```python
Agency = client.query('''
SELECT DISTINCT(Agency) 
FROM (SELECT * FROM `rock-wonder-317907.h8_datasets.travel_insurance`
WHERE Agency_Type = 'Travel Agency')
''').to_dataframe()
```

```python
Agency
```

Pada tabel diatas, dapat kita lihat ada 10 agensi travel yang klaim asuransi. Ada agensi EPX, CWT, KML, LWC, CBH, CSR, CCR, ADM, TTW, dan TST. Agensi travel tersebut dikategorikan berdasarkan kolom Agency_Type dengan spesifik 'Travel Agency'.

3. Tipe agensi apa yang paling banyak yang klaim asuransinya diterima?

```python
Agency_Type = client.query('''
SELECT DISTINCT(Agency_Type), count(Claim) AS Total_Claim 
FROM `rock-wonder-317907.h8_datasets.travel_insurance`
WHERE Claim = True
GROUP BY Agency_Type
ORDER BY Total_Claim DESC
''').to_dataframe()
```

```python
Agency_Type
```

Pada tabel diatas, dapat kita lihat tipe agensi apa yang paling banyak claim asuransinya diterima. Yang pertama ada tipe agensi Airlines dengan total claim yang diterima sebanyak 591, kedua ada tipe agensi Travel Agency dengan total claim yang diterima sebanyak 336.

4. Destinasi negara mana yang paling banyak diterima klaim asuransinya?

```python
Destination = client.query('''
SELECT DISTINCT(Destination), count(Claim) AS Total_Claim 
FROM `rock-wonder-317907.h8_datasets.travel_insurance`
WHERE Claim = True
GROUP BY Destination
ORDER BY Total_Claim DESC
''').to_dataframe()
```

```python
Destination
```

Pada tabel diatas, dapat kita lihat terdapat 41 negara yang klaimnya diterima. Kemudian kita akan lihat negara mana yang klaim diterimanya paling tinggi. Yang pertama ada negara Singapore dengan total klaim yang diterima sebanyak 562, kedua ada negara China dengan total klaim yang diterima sebanyak 46, ketiga ada negara Aurstralia dengan total klaim yang diterima sebanyak 38.

5. Berapa nilai minimum, maksimum, dan rata-rata usia yang pengajuan klaim asuransinya diterima?

```python
Age = client.query('''
SELECT MIN(Age) as Min_Age, MAX(Age) AS Max_Age, AVG(Age) AS AVG_Age
FROM `rock-wonder-317907.h8_datasets.travel_insurance`
WHERE Claim = True
''').to_dataframe()
```

```python
Age
```

Pada tabel diatas, dapat kita lihat minimum usia, maximum usia dan usia rata-rata yang klaim asuransinya diterima. Minimun usia yang klaim asuransinya diterima pada usia 14 tahun, kemudian maximum usia yang klaim asuransinya diterima pada usia 118 tahun, terakhir rata-rata usia yang klaimnya diterima berada pada usia 38 tahun. Umur 38 tahun memang sudah umur yang memasuki banyak penyakit, sehingga range umur tersebut banyak yang klaim asuransi.

6. Jika agen asuransi hanya mengurusi klaim yang tujuan negaranya Singapura, Thailand, dan Malaysia, berapa total komisi yang didapat dari klaim yang diterima? (anggap komisi dalam USD)

```python
Commision = client.query('''
SELECT SUM(Commision__in_value_) as Commision_USD
FROM `rock-wonder-317907.h8_datasets.travel_insurance`
WHERE Claim = True AND Destination = 'SINGAPORE' OR Destination = 'THAILAND' OR Destination = 'MALAYSIA'
''').to_dataframe()
```

```python
Commision
```

Dari tabel diatas, kita dapat mengetahui berapa komisi yang didapat dari klaim asuransi yang diterima dengan kondisi agen asuransu hanya mengurusi klaim yang tujuan negaranya Singapore, Thailand, dan Malaysia. Kita dapat lihat komisi yang didapat oleh agen asuransi yang hanya mengurusi negara Singapore, Thailand dan Malaysia sebesar 85,688 USD. Komisi tersebut sangat besar karena negara Singapore merupakan negara dengan peringkat 1 yang total klaimnya diterima.
